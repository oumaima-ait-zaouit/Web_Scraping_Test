```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

```python
url=requests.get(f"https://www.jumia.ma/telephone-tablette/?q=phones#catalog-listing")
```

```python
soup=BeautifulSoup(url.text,'lxml')
```

```python
page = 1
img = []
mall_store = []
name = []
current_price = []
old_price = []
discount = []


product = soup.find_all('article', class_="prd _fb col c-prd")

for item in product:
    imgs_element = item.find("img", class_="img")
    imgs = imgs_element.get('src') if imgs_element else None

    mall_stores_element = item.find("div", class_="bdg _mall _xs")
    mall_stores = mall_stores_element.text if mall_stores_element else None

    names_element = item.find("h3", class_="name")
    names = names_element.text if names_element else None

    current_prices_element = item.find("div", class_="prc")
    current_prices = current_prices_element.text if current_prices_element else None

    old_prices_element = item.find("div", class_="old")
    old_prices = old_prices_element.text if old_prices_element else None

    discounts_element = item.find("div", class_="bdg _dsct _sm")
    discounts = discounts_element.string if discounts_element else None


    img.append(imgs)
    mall_store.append(mall_stores)
    name.append(names)
    current_price.append(current_prices)
    old_price.append(old_prices)
    discount.append(discounts)

    if page == 50:
        isHaveNextPage = False
    page += 1
```
